# import library

```python
import pandas as pd
from matplotlib import pyplot as plt
```


# Membaca data

```python
df = pd.read_csv("Bengaluru_House_Data.csv")
df
```

```python
df.head()
```

# Visualisasi Data

# Grafik Batang
Dari grafik batang dibawah menunjukan Rumah Berbalkon dengan banyak harga

```python
x = list(df["balcony"])
y = list(df["price"])

plt.xlabel("Balkon")
plt.ylabel("Harga")
plt.title("Grafik Batang")
plt.bar(x,y,color="red")
plt.show()
```

# Grafik pie
Grafik dibawah menunjukan banyak type area yang tersedia

```python
data_pie = df["area_type"].value_counts()
data_pie.head()
```

```python
data_pie = df["availability"].value_counts().rename_axis("area_type")
data_pie.head()
```

```python
data_pie = df["area_type"].value_counts().rename_axis("area_type").reset_index(name="id_count")
data_pie.head()
```

```python
plt.title("Grafik Pie")
plt.pie(data_pie["id_count"],labels=data_pie["area_type"],startangle=90,autopct="%.2f%%",shadow=True,explode=(0,0,0,0.05))
plt.show()
```

# Grafik garis
grafik dibawah menunjukan type area dengan banyak balkon

```python
x = list(df["area_type"])
y = list(df["balcony"])
plt.title("Grafik Garis")
plt.plot(x,y)
plt.show()
```
